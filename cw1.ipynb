# 95% confidence intervals for time distribution

```r
errorX <-  qt(0.975,df=length(fish1$V1)-1)*sd(fish1$V1)/sqrt(length(fish1$V1))
leftX <- mean(fish1$V1)-errorX
rightX <- mean(fish1$V1)+errorX
leftX
rightX
```

# 95% confidence intervals for weight distribution

```r
errorY <-  qt(0.975,df=length(fish1$V2)-1)*sd(fish1$V2)/sqrt(length(fish1$V2))
leftY <- mean(fish1$V2)-errorY
rightY <- mean(fish1$V2)+errorY
leftY
rightY
```
